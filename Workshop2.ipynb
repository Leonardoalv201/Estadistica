```python
from symbulate import *
#'exec(%matplotlib inline)' 
import matplotlib

import matplotlib.pyplot as plt
```

```python
# clear the plot space:
plt.clf()
# Define the probability type of function as uniform (probability space)
P = Uniform(0,40)
# Define X as a random variable of the type P (uniform)
X =RV(P)
# Simulate the values of X and save the in the x array: 
x=X.sim(10000)
x.plot()
plt.show()
```

EN EL HISTOGRAMA PODEMOS VER QUE SE GENERARON DATOS DESDE 0 HASTA 40, Y QUE SE DISTRUBUYERON DE FORMA EQUITATIVA, ESTO QUIERE DECIR QUE NO TENEMOS MUCHOS DE UNO SOLO O MUY POCOS DE ALGUNO

```python
x.mean()
```

```python
x.std()
```

PODEMOS VER QUE LA MEDIA Y LA DESFVIACION ESTANDAR SON MUY SIMILARES A LAS TEÓRICAS, Y ESTO QUIERE DECIR QUE EL ALGORITMO RANDOM ES BUENO 

```python
# Define the probability space as a vector of 25 uniform random numbers:
P = Uniform(0,40) ** 25
# Define XVECTOR as a vector of random variables with 25 random values 
XVECTOR =RV(P)
# Simulate 10,000 vectors to get a matrix  
xmatrix=XVECTOR.sim(10000)
```

```python
xmatrix
```

```python
# The 2nd parameter is the mathematical operation applied to each row of the random vector
# Simulate 10,000 means of each random vector of 25 uniform distributed numbers: 
xmean = RV(P,mean).sim(10000)
xmean
```

```python
plt.clf()
xmean.plot()
plt.show()
```

EN ESTE CASO, TENEMOS UN HISTOGRAMA QUE MUESTRA UNA CAMPANA DE GAUSS, A DIFERENCIA DEL ANTERIOR, ESTO SE DEBE A QUE SUS VALORES SON PROMEDIOS DE 20 DATOS RANDOM ENTRE 0 Y 40, POR LO QUE EL PROMEDIO SE ACERCARIA A 20 PERO TENIENDO PEQUEÑAS DESVIACIONES 

```python
print(xmean.mean())
print(x.mean())
```

```python
print(xmean.std())
print(x.std())
```

```python
# clear the plot space:
plt.clf()
# Define the probability type of function as uniform (probability space)
P = Normal(mean=20, sd=10)
# Define Y as a random variable of the type P (normal)
Y =RV(P)
# Simulate 10,000 values of Y and save the in the x array: 
y=Y.sim(10000)
y.plot()
plt.show()
```

PODEMOS VER UNA CAMPANA DE GAUSS, LA CUAL TIENDE A 20, PERO DEBIDO A SU DESVIACIÓN, LLEGA A TENER VALORES DE HASTA -20 Y 60 Y LA DENSIDAD DE QUE OCURRAN LOS DATOS ES MENOR A LA DE CUANDO OBTUVIMOS CON LA MEDIA

```python
y.std()
```

LA MEDIA, Y DESVIACION ESTANDAR SON MUY SIMILARES A LAS DADAS, YA QUE LA MEDIA DADA ES DE 20 Y LA OBTENIDA ES DE 20+-.2, AL IGUAL QUE LA DESVIACIÓN

```python
# Define the probability space P as a vector of 25 normal random numbers:
P = Normal(mean=20, sd=10) ** 25
# Define RANDOMVECTOR as a vector of 25 random variables in the Probability space P 
RANDOMVECTOR = RV(P)
# Simulate 10,000 random vectors with this distribution and save the in ymatrix: 
ymatrix=RANDOMVECTOR.sim(10000)
```

```python
ymatrix
```

```python
# The 2nd parameter is the mathematical operation applied to each row of the random vector
# Simulate 10,000 means of each random vector of 25 uniform distributed numbers: 
ymean = RV(P,mean).sim(10000)
ymean
```

```python
plt.clf()
ymean.plot()
plt.show()
```

PODEMOS VER QUE LA PROBABILIDAD DE TENER 20 ES MAYOR, Y QUE LA DESVIACION DISMINUYO, YA QUE AHORA SOLO ES DE 12 A 28

```python
print(ymean.std())
print(y.std())
```

```python
import numpy as np
import math
import pandas as pd
import pandas_datareader as pdr
import matplotlib
import matplotlib.pyplot as plt

MSFT = pdr.get_data_yahoo('MSFT',
start="01/01/2017", interval="m")
```

# 1.
VAMOS A USAR COMO NUESTRA VARIABLE DE ESTUDIO LA DIVISION ENTRE EL VALOR DE ADJ. CLOSE DE CADA MES ENTRE EL DEL MES ANTERIOR, PARA CONOCER EL CAMBIO 

```python
MSFT["r"] = np.log(MSFT['Adj Close']) - np.log(MSFT['Adj Close'].shift(1))
print(MSFT["r"])
```

# 2.
##Null hypothesis:
El promedio del cambio en el valor mensual en las acciones de Microsoft es igual a 0
##Alternative hypothesis:
El promedio del cambio en el valor mensual en las acciones de Microsoft es mayor a 0

```python
R_mean = MSFT["r"].mean()
print(R_mean)
```

#3.

```python
R_std = MSFT["r"].std()
print(R_std)
```

```python
R_std_grouped = R_std / math.sqrt(69)
R_std_grouped
```

#4.


```python
plt.hist(MSFT["r"], bins = 10)
plt.show()
```

```python
t_stadistic = R_mean / R_std_grouped
t_stadistic
```

PODEMOS VER QUE TENEMOS UNA T_STATISTIC MAYOR A 3, POR LO QUE TEMOS UNA DESVIACIÓN DE MÁS DE 99%, POR LO QUE LA HIPOTESIS NULA PUEDE SER DESCARTADA



#5.

```python
MIN = R_mean - 2 * R_std_grouped
MAX = R_mean + 2 * R_std_grouped
print(MIN, MAX)
```
