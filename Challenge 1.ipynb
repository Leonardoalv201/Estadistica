```python
BTC["R"] = (BTC["Close"] / BTC["Close"].shift(1)) - 1
print(BTC["R"])
```

```python
BTC["r"] = np.log(BTC['Adj Close']) - np.log(BTC['Adj Close'].shift(1))
BTCR = BTC[['R','r']].copy()
```

```python
BTC= BTC.dropna()
```

```python
sumret = BTC["R"].describe()
sumret
```

```python
print("The daily average return of Bitcoin is ", BTC["r"].mean())
print("The daily variance of Bitcoin is ", BTC["r"].var())
print("The daily standard deviation (volatility) of Bitcoin return is ", BTC["r"].std())
print("The daily median return of Bitcoin is ",BTC["r"].median())
```

```python
r_bitcoin = pd.DataFrame(BTC[["R"]])
hist = r_bitcoin.plot.hist(bins = 12, alpha = .5, title = "Histogram of daily Bitcoin Returns")
```

PODEMOS OBSERVAR DE ESTE HISTOGRAMA, QUE EL BITCOIN TIENDE MÁS A AUMENTAR SU VALOR. AUNQUE DE IGUAL FORMA, SE PUEDE VER QUE EXISTEN REGISTROS DONDE TIENDE A DISMINUIR MÁS ABRUPTAMENTE QUE A SUBIR, POR LO QUE SI LLEGASE A DISMINUIR, TE AFECTARIA MÁS DE LO QUE TE BENEFICIARIA CUANDO SUBE, AUNQUE ES MENOS PROBABLE QUE OCURRA.

```python
R_mean=BTC["R"].mean()
R_std=BTC["R"].std()
R_size=len(BTC)
```

```python
from matplotlib import pyplot

bins=200

R_sim=np.random.normal(loc=R_mean, scale=R_std, size=R_size)
pyplot.hist(R_sim, bins, alpha=0.5, label="Simulated")
pyplot.hist(r_bitcoin.to_numpy(), bins, alpha=0.5, label="Real")
pyplot.legend(loc="upper right")
pyplot.show()
```

EN ESTE CASO, DONDE ESTAMOS USANDO UNA SIMULACIÓN CON NÚMEROS RANDOM, PODEMOS VER QUE ESTOS SE DISTRIBUYEN DE DORMA MÁS EQUITATIVA, CUMPLIENDO MÁS CON LA FORMA DE UNA CAMPANA DE GAUSS, TENIENDO UNA DISPERCIÓN MENOR Y MENOS DATOS EN LOS EXTREMOS
